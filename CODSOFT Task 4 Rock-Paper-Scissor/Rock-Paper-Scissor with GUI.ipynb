```python
from tkinter import *
from random import randint
from tkinter import ttk

root = Tk()
root.title('Rock_Paper_Scissor')
root.iconbitmap("")
root.geometry("800x750")
root.config(bg="white")

rock = PhotoImage(file="Images/rock.png")
paper = PhotoImage(file="Images/paper.png")
scissors = PhotoImage(file="Images/scissors.png")

image_list = [rock, paper, scissors]

pick_number = randint(0, 2)

image_label = Label(root, image=image_list[pick_number], bd=0)
image_label.pack(pady=5)

user_choice = ttk.Combobox(root, value=("Rock", "Paper", "Scissors"))
user_choice.current(0)
user_choice.pack(pady=5)

win_lose_label = Label(root, text="", font=("Helvetica", 18), bg="white")
win_lose_label.pack(pady=5)


def spin():
    #user_choice_value = None
    if user_choice.get() == "Rock":
        user_choice_value = 0
    elif user_choice.get() == "Paper":
        user_choice_value = 1
    elif user_choice.get() == "Scissors":
        user_choice_value = 2

    pick_number = randint(0, 2)
    image_label.config(image=image_list[pick_number])
    image_label.image = image_list[pick_number]

    if user_choice_value == 0:
        if pick_number == 0:
            win_lose_label.config(text="Its a tie")
        elif pick_number == 1:
            win_lose_label.config(text="Paper beats rock! You Lose")
        elif pick_number == 2:
            win_lose_label.config(text="Rock beats Scissors! You Won")
    elif user_choice_value == 1:
        if pick_number == 1:
            win_lose_label.config(text="Its a tie")
        elif pick_number == 0:
            win_lose_label.config(text="Paper beats rock! You Won")
        elif pick_number == 2:
            win_lose_label.config(text="Scissors beats paper! You Lose")
    elif user_choice_value == 2:
        if pick_number == 2:
            win_lose_label.config(text="Its a tie")
        elif pick_number == 0:
            win_lose_label.config(text="Rock beats Scissors! You Lose")
        elif pick_number == 1:
            win_lose_label.config(text="Scissors beats paper! You Won")


spin_button = Button(root, text="Click & Win", command=spin)
spin_button.pack(pady=20)

root.mainloop()
```
