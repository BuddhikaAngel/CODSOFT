```python
from asyncore import write
import sys
import csv

def add(i):
    with open('data.csv', 'a+', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(i)
        
#add(['Nimal', '+94789950632', 'nimal@gmail.com', 'Negombo'])
#add(['Kamal', '+947994506342', 'kamal@gmail.com', 'Negombo'])

def view():
    data = []
    with open('data.csv') as file:
        reader = csv.reader(file)
        for row in reader:
            data.append(row)
            
    print(data)
    return data
view()

def remove(i):    
    def save(j):
        with open('data.csv', 'w', newline='') as file:
            writer = csv.writer(file)
            writer.writerows(j)
            
    new_list = []
    telephone = i
    
    with open('data.csv','r') as file:
        reader = csv.reader(file)
        for row in reader:
            new_list.append(row)
            
            for element in row:
                if element == telephone:
                    new_list.remove(row)
    save(new_list)
    
def update(i):
    def update_newlist(j):
        with open('data.csv', 'w', newline='') as file:
            writer = csv.writer(file)
            writer.writerows(j)
    new_list = []
    telephone = i[0]
    
    with open('data.csv', 'r') as file:
        reader = csv.reader(file)
        for row in reader:
            new_list.append(row)
            for element in row:
                if element == telephone:
                    name = i[1]
                    telephone = i[2]
                    email = i[3]
                    address = i[4]
                        
                    data = [name, telephone, email, address]
                    index = new_list.index(row)
                    new_list[index] = data
    update_newlist(new_list)
    
    
def search(i):
    data = []
    telephone = i
    
    with open('data.csv','r') as file:
        reader = csv.reader(file)
        for row in reader:
            for element in row:
                if element == telephone:
                    data.append(row)
    print(data)
    return data
```

```python
from tkinter import *
from tkinter import ttk
from tkinter import messagebox

window = Tk()
window.title("")
window.geometry("485x450")
window.config(bg="Black")
window.resizable(width=FALSE, height=FALSE)
    
frame_up = Frame(window, width=500, height=50, bg="black")
frame_up.grid(row=0, column=0, padx=0, pady=1)

frame_down = Frame(window, width=500, height=150, bg="white")
frame_down.grid(row=1, column=0, padx=0, pady=1)

frame_table = Frame(window, width=500, height=100, bg="white", relief="flat")
frame_table.grid(row=2, columnspan=2, padx=10, pady=1, stick=NW)

#function
def show():
    global tree
    
    listheader = ['Name', 'Telephone', 'Email','Address']
    
    demo_list = view()
    
    #demo_list = [['Nimasha', '+94769950635', 'buddhikaangel@gmail.com', 'Kadawatha']]
    
    tree = ttk.Treeview(frame_table, selectmode="extended", columns=listheader, show="headings")
    
    vb = ttk.Scrollbar(frame_table, orient="vertical", command=tree.yview)
    hb = ttk.Scrollbar(frame_table, orient="horizontal", command=tree.xview)
    
    tree.configure(yscrollcommand=vb.set, xscrollcommand=hb.set)
    
    tree.grid(column=0, row=0, sticky='nsew')
    vb.grid(column=1, row=0, sticky='ns')
    hb.grid(column=0, row=1, sticky='ew')
    
    tree.heading(0, text='Name', anchor=NW)
    tree.heading(1, text='Telephone', anchor=NW)
    tree.heading(2, text='Email', anchor=NW)
    tree.heading(3, text='Address', anchor=NW)
    
    tree.column(0, width=85, anchor='nw')
    tree.column(1, width=100, anchor='nw')
    tree.column(2, width=180, anchor='nw')
    tree.column(3, width=85, anchor='nw')
    
    for item in demo_list:
        tree.insert('', 'end', values=item)
show()

def insert():
    Name = e_name.get()
    Telephone = e_telephone.get()
    Email = e_email.get()
    Address = e_address.get()
    
    data = [Name, Telephone, Email, Address]
    if Name == '' or Telephone == '' or Email == '' or Address == '':
        messagebox.showwarning('data', 'Please fill in all fields')
    else:
        add(data)
        messagebox.showinfo('data', 'Data added successfully')
        e_name.delete(0, 'end')
        e_telephone.delete(0, 'end')
        e_email.delete(0, 'end')
        e_address.delete(0, 'end')    
        
    show()
    
def to_update():
    try:
        tree_data = tree.focus()
        tree_dictionary = tree.item(tree_data)
        tree_list = tree_dictionary['values']
        
        Name = str(tree_list[0])
        Telephone = str(tree_list[1])
        Email = str(tree_list[2])
        Address = str(tree_list[3])
        
        e_name.insert(0, Name)
        e_telephone.insert(0, Telephone)
        e_email.insert(0, Email)
        e_address.insert(0, Address)
        
        def confirm():
            new_name = e_name.get()
            new_telephone = e_telephone.get()
            new_email = e_email.get()
            new_address = e_address.get()
            
            data = [new_telephone, new_name, new_telephone, new_email, new_address]
            
            update(data)
            
            messagebox.showinfo('Sucsess', 'Data updated Successfully')
            
            e_name.delete(0, 'end')
            e_telephone.delete(0, 'end')
            e_email.delete(0, 'end')
            e_address.delete(0, 'end')
            
            for widget in frame_table.winfo_children():
                widget.destroy()
                
            b_confirm.destroy()
            
            show()
            
        b_confirm = Button(frame_down, width=12, text="Click to Update", height=1, bg="black", fg="white", font=('Ivy 10'), command=confirm)
        b_confirm.place(x=275, y=80)
        
    except IndexError:
            messagebox.showerror('Error', 'Select one from the list')
            
def to_remove():
    try:
        tree_data = tree.focus()
        tree_dictionary = tree.item(tree_data)
        tree_list = tree_dictionary['values']
        tree_telephone = str(tree_list[2])
        
        remove(tree_telephone)
        
        messagebox.showinfo('Success','Data has been deleted successfully')
        
        for widget in frame_table.winfo_children():
            widget.destroy()
        show()   
    except IndexError:
            messagebox.showerror('Error', 'Select one from the list')
        
def to_search():
    telephone = e_search.get()
    
    data = search(telephone)
    
    def delete_command():
        tree.delete(*tree.get_children())
    
    delete_command()
    
    for item in data:
        tree.insert('', 'end', values=item)
    e_search.delete(0, 'end')

    
app_name = Label(frame_up, text="Phone Book", height=1, font=('Verdana 17 bold'), bg="black", fg="white")
app_name.place(x=5, y=5)

l_name = Label(frame_down, text="Name *", width=20, height=1, font=('Ivy 10'), bg="white", anchor=NW)
l_name.place(x=10, y=20)

e_name = Entry(frame_down, width=25, justify='left', highlightthickness=1, relief='solid')
e_name.place(x=80, y=20)

#Telephone
l_telephone = Label(frame_down, text="Telephone *", width=20, height=1, font=('Ivy 10'), bg="white", anchor=NW)
l_telephone.place(x=10, y=50)

e_telephone = Entry(frame_down, width=25, justify='left', highlightthickness=1, relief='solid')
e_telephone.place(x=80, y=50)

#Email
l_email = Label(frame_down, text="Email *", width=20, height=1, font=('Ivy 10'), bg="white", anchor=NW)
l_email.place(x=10, y=80)

e_email = Entry(frame_down, width=25, justify='left', highlightthickness=1, relief='solid')
e_email.place(x=80, y=80)

#Address
l_address = Label(frame_down, text="Address *", width=20, height=1, font=('Ivy 10'), bg="white", anchor=NW)
l_address.place(x=10, y=110)

e_address = Entry(frame_down, width=25, justify='left', highlightthickness=1, relief='solid')
e_address.place(x=80, y=110)

#search button
b_search = Button(frame_down, text="Search", height=1, bg="black", fg="white", font=('Ivy 10'), command=to_search)
b_search.place(x=290, y=20)

e_search = Entry(frame_down, width=20, justify='left', highlightthickness=1, relief='solid')
e_search.place(x=345, y=20)

b_view = Button(frame_down, width=10, text="View", height=1, bg="black", fg="white", font=('Ivy 10'), command=show)
b_view.place(x=290, y=50)

b_add = Button(frame_down, width=10, text="Add", height=1, bg="black", fg="white", font=('Ivy 10'), command=insert)
b_add.place(x=390, y=50)

b_update = Button(frame_down, width=10, text="Update", height=1, bg="black", fg="white", font=('Ivy 10'), command=to_update)
b_update.place(x=390, y=80)

b_delete = Button(frame_down, width=10, text="Delete", height=1, bg="black", fg="white", font=('Ivy 10'), command=to_remove)
b_delete.place(x=390, y=110)

window.mainloop()
```
